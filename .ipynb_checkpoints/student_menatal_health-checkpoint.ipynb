```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif

data_path = 'students_mental_health_survey.csv'
df = pd.read_csv(data_path)

display(df)
```

```python
null_values = df.isnull().sum()

# Display the null values count for each column
print("Null values count for each column:")
print(null_values)
```

```python
df_cleaned_rows = df.dropna(how='any')



null_values = df_cleaned_rows.isnull().sum()

# Display the null values count for each column
print("Null values count for each column:")
print(null_values)
```

```python

# Data Preprocessing

# Feature Engineering

# 1. Create Age Groups based on Age
df['Age'] = pd.cut(df['Age'], bins=[0, 19, 29, 39, float('inf')],
                          labels=['Teen', 'Young Adult', 'Adult', 'Senior'])

# 2. Categorize Credit Load into Low, Medium, and High
df['Semester_Credit_Load'] = pd.cut(df['Semester_Credit_Load'], bins=[15, 20, 25, float('inf')],
                                      labels=['Low', 'Medium', 'High'])

# 3. Categorize CGPA into Low, Medium, High, and Excellent
df['CGPA'] = pd.cut(df['CGPA'], bins=[0, 3, 3.5, 4, float('inf')],
                              labels=['Low', 'Medium', 'High', 'Excellent'])
```

```python
from sklearn.preprocessing import LabelEncoder
df = df_cleaned_rows


# Handle missing values in numerical columns by filling with the mean
numerical_columns = ['Age', 'CGPA', 'Stress_Level', 'Depression_Score', 'Anxiety_Score', 'Financial_Stress', 'Semester_Credit_Load']
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())


# Handle missing values in categorical columns by filling with the mode (most frequent value)
categorical_columns = [
    'Course', 'Gender', 'Sleep_Quality', 'Physical_Activity', 'Diet_Quality', 'Social_Support',
    'Relationship_Status', 'Substance_Use', 'Counseling_Service_Use', 'Family_History',
    'Chronic_Illness', 'Extracurricular_Involvement', 'Residence_Type',
    'Age', 'Credit_Load', 'CGPA']

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
    

    
    
    
Course = LabelEncoder()
df['Course'] = Course.fit_transform(df['Course'])
print(df["Course"].unique())


Gender = LabelEncoder()
df['Gender'] =  Gender.fit_transform(df['Gender'])
print(df["Gender"].unique())

Sleep_Quality = LabelEncoder()
df['Sleep_Quality'] = Sleep_Quality.fit_transform(df['Sleep_Quality'])
print(df["Sleep_Quality"].unique())

Physical_Activity= LabelEncoder()
df['Physical_Activity'] = Physical_Activity.fit_transform(df['Physical_Activity'])
df["Physical_Activity"].unique()

Diet_Quality= LabelEncoder()
df['Diet_Quality'] = Diet_Quality.fit_transform(df['Diet_Quality'])
df["Diet_Quality"].unique()

Social_Support = LabelEncoder()
df['Social_Support'] = Social_Support.fit_transform(df['Social_Support'])
df["Social_Support"].unique()

Relationship_Status = LabelEncoder()
df['Relationship_Status'] = Relationship_Status.fit_transform(df['Relationship_Status'])
df["Relationship_Status"].unique()


Substance_Use= LabelEncoder()
df['Substance_Use'] = Substance_Use.fit_transform(df['Substance_Use'])
df["Substance_Use"].unique()

Counseling_Service_Use= LabelEncoder()
df['Counseling_Service_Use'] = Counseling_Service_Use.fit_transform(df['Counseling_Service_Use'])
df["Counseling_Service_Use"].unique()

Family_History= LabelEncoder()
df['Family_History'] = Family_History.fit_transform(df['Family_History'])
df["Family_History"].unique()

Chronic_Illness= LabelEncoder()
df['Chronic_Illness'] = Chronic_Illness.fit_transform(df['Chronic_Illness'])
df["Chronic_Illness"].unique()

Extracurricular_Involvement= LabelEncoder()
df['Extracurricular_Involvement'] =  Extracurricular_Involvement.fit_transform(df['Extracurricular_Involvement'])
df["Extracurricular_Involvement"].unique()

Residence_Type = LabelEncoder()
df['Residence_Type'] = Residence_Type .fit_transform(df['Residence_Type'])
df["Residence_Type"].unique()


# Handle Outliers

# Calculate summary statistics for the dataset
summary = df.describe()
Q1 = summary.loc['25%']
Q3 = summary.loc['75%']
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and remove outliers using the IQR method
is_outlier = ((df < lower_bound) | (df > upper_bound)).any(axis=1)
df = df[~is_outlier]

# Map Stress_Level to binary values

# Map stress level values to binary categories
stress_level_mapping = {0: 0, 1: 1, 2: 1, 3: 2, 4: 2, 5: 3}
data['Stress_Level'] = data['Stress_Level'].replace(stress_level_mapping)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Load the dataset into a Pandas DataFrame
# Replace 'your_dataset.csv' with the actual file path or dataset loading mechanism

# Separate features (X) and target (y)
X = df.drop('Stress_Level', axis=1)  # Features (excluding the target column)
y = df['Stress_Level']  # Target column

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=8)

# Train the Decision Tree Classifier
# You can customize the parameters of the DecisionTreeClassifier as needed
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)


# Predict the target values for the testing set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Classification Report:')
print(classification_report(y_test, y_pred))


print('Accuracy:', accuracy)
```
